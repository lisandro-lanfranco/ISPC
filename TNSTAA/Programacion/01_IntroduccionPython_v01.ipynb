## Operadores Matemáticos

| **Operador** | **Operacion**   | **Ejemplo**   |
|--------------|-----------------|---------------|
| +            | Suma            | 2 + 3 = 5     |
| -            | Resta           | 5 - 2 = 3     |
| *            | Multiplicación  | 2 * 5 = 10    |
| /            | División        | 10 / 3 = 3.33 |
| //           | División Entera | 10 // 3 = 3   |
| %            | Resto           | 10 % 3 = 1    |
| **           | Exponente       | 2 ** 3 = 8    |

Ordenados desde la menor al mayor orden de evaluación.

```python
21+3//3
```
